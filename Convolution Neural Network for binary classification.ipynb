```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Activation, Conv2D, MaxPooling2D
import pickle
#convolution --> pooling --> convolution -->pooling
#simplification of features from convolution window to smaller representation of the image
#in an image it can find edges and lines, these features can then be highlighted through convolutin and pooling

X = pickle.load(open("X.pickle", "rb"))
y = pickle.load(open("Y.pickle", "rb"))

#normalizing our image data
X = X/255.0
#print(X)
#print(y)
model = Sequential()
model.add(Conv2D(64, (3,3), input_shape = X.shape[1:]))# some size,window size, input size
#activation or pooling
model.add(Activation ('relu'))
model.add(MaxPooling2D(pool_size= (2,2)))

model.add(Conv2D(64, (3,3)))# some size,window size
#activation or pooling
model.add(Activation ('relu'))
model.add(MaxPooling2D(pool_size= (2,2)))

model.add(Flatten())
#Dense needs 1D data
model.add(Dense(64))
#output layer
model.add(Dense(1))
model.add(Activation('sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.fit(X, y, batch_size=32, epochs= 3, validation_split=0.1) #batch size will be discussed in next one
```
